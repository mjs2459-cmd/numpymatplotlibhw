```python
# Creating and Manipulating Arrays

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create two 2D arrays representing coordinates x, y on the cartesian plan

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
```

```python
X, Y = np.meshgrid(x, y)
```

```python
# 1.2. Visualize each 2D array using pcolormesh

plt.figure()
plt.pcolormesh(X, Y, X)
plt.colorbar()
plt.title("X values")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
plt.figure()
plt.pcolormesh(X, Y, Y)
plt.colorbar()
plt.title("Y values")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 1.3 From your cartesian coordinates, create polar coordinates 

R = np.sqrt(X**2 + Y**2)
Theta = np.arctan2(Y, X)
```

```python
# 1.4 Visualize R

plt.figure()
plt.pcolormesh(X, Y, R)
plt.colorbar()
plt.title("Radius (R)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 1.4 Visualize Theta

plt.figure()
plt.pcolormesh(X, Y, Theta)
plt.colorbar()
plt.title("Angle (Theta)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 1.5 Calculate the quanity

f = np.cos(4*R)**2 + np.sin(4*Theta)**2
```

```python
plt.figure()
plt.pcolormesh(X, Y, f)
plt.colorbar()
plt.title("f = cos^2(4r) + sin^2(4φ)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import pooch
url = "https://www.ldeo.columbia.edu/~danielmw/float_data.zip"
files = pooch.retrieve(url, processor=pooch.Unzip(), known_hash="2a703c720302c682f1662181d329c9f22f9f10e1539dc2d6082160a469165009")
files
```

```python
for f in files:
    print(f)
```

```python
# 2.1 Load each data file as a numpy array.

import numpy as np

T = np.load(files[1])
S = np.load(files[2])
P = np.load(files[3])
level = np.load(files[4]) 
lon = np.load(files[5])
lat = np.load(files[6])
date = np.load(files[0])
```

```python
# 2.2 Examine the shapes of T, S and P compared to lon, lat, date and level

print("T shape:", T.shape)
print("S shape:", S.shape)
print("P shape:", P.shape)
print("level shape:", level.shape)
print("lon shape:", lon.shape)
print("lat shape:", lat.shape)
```

```python
# Answer to 2.2: T, S, and P share two dimensions: 78 depth levels (matching level) and 75 float profiles (matching lon and lat).
```

```python
# 2.3 Make a plot for each column of data in T, S and P (three plots)

import matplotlib.pyplot as plt

plt.figure()
plt.plot(T, level)
plt.gca().invert_yaxis()
plt.title("Temperature Profiles")
plt.xlabel("Temperature")
plt.ylabel("Depth")
plt.show()
```

```python
plt.figure()
plt.plot(S, level)
plt.gca().invert_yaxis()
plt.title("Salinity Profiles")
plt.xlabel("Salinity")
plt.ylabel("Depth")
plt.show()
```

```python
plt.figure()
plt.plot(P, level)
plt.gca().invert_yaxis()
plt.title("Pressure Profiles")
plt.xlabel("Pressure")
plt.ylabel("Depth")
plt.show()
```

```python
axis=1
```

```python
# 2.4 Compute the mean and standard deviation of each of T, S and P at each depth in level

T_mean = np.mean(T, axis=1)
T_std = np.std(T, axis=1)

S_mean = np.mean(S, axis=1)
S_std = np.std(S, axis=1)

P_mean = np.mean(P, axis=1)
P_std = np.std(P, axis=1)
```

```python
# 2.5 Now make three similar plot, but show only the mean T, S and P at each depth. Show error bars on each plot using the standard deviations

plt.figure()
plt.errorbar(T_mean, level, xerr=T_std)
plt.gca().invert_yaxis()
plt.title("Mean Temperature with Std Dev")
plt.xlabel("Temperature")
plt.ylabel("Depth")
plt.show()
```

```python
plt.figure()
plt.errorbar(S_mean, level, xerr=S_std)
plt.gca().invert_yaxis()
plt.title("Mean Salinity with Std Dev")
plt.xlabel("Salinity")
plt.ylabel("Depth")
plt.show()
```

```python
plt.figure()
plt.errorbar(P_mean, level, xerr=P_std)
plt.gca().invert_yaxis()
plt.title("Mean Pressure with Std Dev")
plt.xlabel("Pressure")
plt.ylabel("Depth")
plt.show()
```

```python
# 2.6 Account For Missing Data

T_mean = np.nanmean(T, axis=1)
T_std = np.nanstd(T, axis=1)

S_mean = np.nanmean(S, axis=1)
S_std = np.nanstd(S, axis=1)

P_mean = np.nanmean(P, axis=1)
P_std = np.nanstd(P, axis=1)
```
